```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a cross-tab of counts
gender_churn = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to percentages row-wise
gender_churn_percent = gender_churn.div(gender_churn.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = gender_churn_percent.plot(kind='bar', stacked=True, figsize=(6,4), colormap='Set2')

# Step 4: Add % labels to each section of bar
for container in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 0 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=10, color='black')

# Final touches
plt.title("Churn by SeniorCitizen (Stacked Bar Chart)", fontsize=13)
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### comparative a greater percentage of people in senior citizen category have churned

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure SeniorCitizen is Yes/No
df['SeniorCitizen'] = df['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})

# List of categorical columns
cat_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
               'PhoneService', 'MultipleLines', 'InternetService',
               'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies',
               'Contract', 'PaperlessBilling', 'PaymentMethod']

# Define subplot grid size
n_cols = 3
n_rows = (len(cat_columns) + n_cols - 1) // n_cols

# Create figure and axes
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()  # Flatten to easily index with a loop

# Plot each countplot in its subplot
for i, col in enumerate(cat_columns):
    sns.countplot(data=df, x=col, hue='Churn', palette='Set2', ax=axes[i])
    axes[i].set_title( f"{col} Count Plot")
    axes[i].tick_params(axis='x', rotation=30)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Hide any unused axes (in case total columns < n_rows * n_cols)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

###### The majority of customers across all categories did not churn However, churn tends to be higher among customers without partners, those with fiber optic internet, no online security or tech support, and those on month-to-month contracts. Services like online backup, device protection, and streaming TV also show a visible correlation with increased churn when not subscribed.
